```python
import os
import torch
import pandas as pd
from sklearn.model_selection import train_test_split

# --- Config ---
root_dir = "/home/free4ky/projects/chest-diseases/data/valid_fixed_embeds_not_normalized_base"           # path to dataset root
df = dataset                         # your pandas dataframe with volumename + labels
save_dir = "preprocessed_val_base"            # where to save stacked files
os.makedirs(save_dir, exist_ok=True)

label_cols = df.columns[1:]

all_tensors = []
all_labels = []
all_names = []

# Traverse files and collect
for root, _, files in os.walk(root_dir):
    for fname in files:
        if fname.endswith(".pt"):
            volname = fname.replace(".nii.pt", ".nii.gz")
            row = df[df['VolumeName'] == volname]
            if row.empty:
                continue  # skip if not in dataframe

            tensor = torch.load(os.path.join(root, fname))
            label = torch.tensor(row[label_cols].values.squeeze(), dtype=torch.float32)

            all_tensors.append(tensor)
            all_labels.append(label)
            all_names.append(volname)

# Stack all into big tensors
X = torch.stack(all_tensors)
Y = torch.stack(all_labels)

print(f"Full dataset shape: {X.shape}, labels shape: {Y.shape}")
```

```python
num_samples = Y.shape[0]

# make zeros of shape [num_samples, 2]
zeros = torch.zeros((num_samples, 2), dtype=Y.dtype, device=Y.device)

# concatenate along the last dimension
new_y = torch.cat([Y, zeros], dim=1)

print(Y.shape)  # [N, 18]
print(new_y.shape)  # [N, 20]
```

```python
# torch.save(X, os.path.join(save_dir, f"validation_data.pt"))
# torch.save(Y, os.path.join(save_dir, f"validation_labels.pt"))
```

```python

# --- Train/Val/Test split (80/10/10) ---
train_idx, val_idx = train_test_split(range(len(X)), test_size=0.1, random_state=42, shuffle=True)
# val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42, shuffle=True)

splits = {
    "train": train_idx,
    "val": val_idx,
    # "test": test_idx
}

for split_name, indices in splits.items():
    split_X = X[indices]
    split_Y = new_y[indices]
    torch.save(split_X, os.path.join(save_dir, f"{split_name}_data.pt"))
    torch.save(split_Y, os.path.join(save_dir, f"{split_name}_labels.pt"))
    print(f"Saved {split_name}: {split_X.shape}, {split_Y.shape}")
```
